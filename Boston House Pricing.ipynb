```python
#importing packages
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
```

```python
#loading dataset
#10% is kept for testing
(train_X, train_Y), (test_X, test_Y) = keras.datasets.boston_housing.load_data(test_split=0.1, seed=10)
```

```python
#splitting training data
#into train and validation set
#25% is kept for validation
train_X, val_X, train_Y, val_Y = train_test_split(train_X, train_Y, test_size=0.25, random_state=10)
```

```python
#data shapes
print(f'Training Data - {train_X.shape}')
print(f'Validation Data - {val_X.shape}')
print(f'Test Data - {test_X.shape}')
```

```python
#creating network
model = keras.Sequential([
    layers.Dense(units=13, activation='relu', input_shape=[13]),
    layers.Dense(units=1)
])
```

```python
#compiling model
model.compile(optimizer='adam', loss='mae', metrics=['accuracy'])

#early stopping
#to avoid overfitting
early_stopping = keras.callbacks.EarlyStopping(patience=15, min_delta=0.001, restore_best_weights=True)
```

```python
#training
history = model.fit(
    train_X, train_Y,
    validation_data = (val_X, val_Y),
    epochs=200,
    batch_size=20,
    callbacks=[early_stopping]
)
```

```python
#learning curves
history_df = pd.DataFrame(history.history)
history_df.loc[:, ['loss', 'val_loss']].plot();
print("Minimum validation loss: {}".format(history_df['val_loss'].min()))
```

```python
#getting predictions
preds = model.predict(test_X)

print(f'Actual Value {test_Y[1]}')
print(f'Prediction {preds[1]}')
```

```python
#model performance
model.evaluate(test_X, test_Y, verbose=2)
```
